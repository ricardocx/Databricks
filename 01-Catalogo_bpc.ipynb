### Este notebook contém os comandos para criar a arquitetura Medalhão.
* Catálogo:
    - bpc
* Esquemas:
    - staging
    - bronze
    - silver
    - gold

```python
%sql

-- Deleta o catálogo bpc se já existir
DROP CATALOG IF EXISTS bpc CASCADE;

-- Cria o catálogo bpc
CREATE CATALOG bpc;

-- Adiciona uma descrição para o catálogo bpc
COMMENT ON CATALOG bpc IS 'Catálogo para o projeto BPC';

-- Deleta o esquema staging se já existir no catálogo bpc
DROP SCHEMA IF EXISTS bpc.staging CASCADE;

-- Cria o esquema staging dentro do catálogo bpc
CREATE SCHEMA bpc.staging;

-- Adiciona uma descrição para o esquema staging
COMMENT ON SCHEMA bpc.staging IS 'Esquema para os volumes de dados de entrada';

-- Deleta o esquema bronze se já existir no catálogo bpc
DROP SCHEMA IF EXISTS bpc.bronze CASCADE;

-- Cria o esquema bronze dentro do catálogo bpc
CREATE SCHEMA bpc.bronze;

-- Adiciona uma descrição para o esquema bronze
COMMENT ON SCHEMA bpc.bronze IS 'Esquema para dados brutos';

-- Deleta o esquema silver se já existir no catálogo bpc
DROP SCHEMA IF EXISTS bpc.silver CASCADE;

-- Cria o esquema silver dentro do catálogo bpc
CREATE SCHEMA bpc.silver;

-- Adiciona uma descrição para o esquema silver
COMMENT ON SCHEMA bpc.silver IS 'Esquema para dados transformados';

-- Deleta o esquema gold se já existir no catálogo bpc
DROP SCHEMA IF EXISTS bpc.gold CASCADE;

-- Cria o esquema gold dentro do catálogo bpc
CREATE SCHEMA bpc.gold;

-- Adiciona uma descrição para o esquema gold
COMMENT ON SCHEMA bpc.gold IS 'Esquema para dados agregados';
```
