### Este notebook contém os comandos para estruturar o esquema staging.
1. Criar o volume imdb.

```python
%sql

-- Remove o volume 'imdb' do esquema 'bpc.staging' se ele existir
DROP VOLUME IF EXISTS bpc.staging.imdb;

-- Cria um novo volume chamado 'imdb' no esquema 'bpc.staging'
CREATE VOLUME bpc.staging.imdb;

-- Adiciona um comentário descritivo ao volume 'imdb' no esquema 'bpc.staging'
COMMENT ON VOLUME bpc.staging.imdb IS 'Volume para armazenar qualquer tipo de dado relacionado ao projeto BPC';
```


2. Importar as bibliotecas

```python
# Importando bibliotecas

import requests

display("Bibliotecas importadas com sucesso.")
```

3. Copiar o arquivo com os dados dos beneficiários.

```python
# URL do arquivo CSV no GitHub
github_url = "https://raw.githubusercontent.com/ricardocx/Databricks/refs/heads/main/dados/Beneficiarios_BPC_2025-10.csv"
# Caminho local onde o arquivo será salvo
local_path = "/Volumes/bpc/staging/imdb/Beneficiarios_BPC_2025-10.csv"

try:
    # Faz o download do arquivo usando requests
    response = requests.get(github_url)
    # Verifica se houve erro na requisição
    response.raise_for_status()
    # Abre o arquivo local para escrita em modo binário
    with open(local_path, "wb") as f:
        # Escreve o conteúdo baixado no arquivo local
        f.write(response.content)    
    display("Arquivo copiado com sucesso.")
except requests.exceptions.RequestException as e:    
    raise RuntimeError(f"Erro ao baixar o arquivo: {e}")
```

4. Copiar o arquivo com os dados descritivos de municípios, unidades da federação e regiões do Brasil.

```python
# URL do arquivo xls no GitHub
github_url = "https://github.com/ricardocx/Databricks/raw/refs/heads/main/dados/Mun_Faixa_de_Fronteira_Cidades_Gemeas_2024.xls"

# Caminho local onde o arquivo será salvo
local_path = "/Volumes/bpc/staging/imdb/Mun_Faixa_de_Fronteira_Cidades_Gemeas_2024.xls"

try:
    # Faz o download do arquivo usando requests
    response = requests.get(github_url)
    # Verifica se houve erro na requisição
    response.raise_for_status()
    # Abre o arquivo local para escrita em modo binário
    with open(local_path, "wb") as f:
        # Escreve o conteúdo baixado no arquivo local
        f.write(response.content)    
    display("Arquivo copiado com sucesso.")
except requests.exceptions.RequestException as e:    
    raise RuntimeError(f"Erro ao baixar o arquivo: {e}")
```
